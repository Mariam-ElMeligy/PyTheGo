# üêº Pandas Practice: DataFrames, Selection, Mapping, and More


```python
import pandas as pd
import numpy as np

# Set seed for reproducibility
np.random.seed(42)

# Sample data
names = ['Alice', 'Bob', 'Charlie', 'Diana', 'Ethan', 'Fiona', 'George', 'Hannah']
departments = ['Sales', 'Marketing', 'HR', 'Tech']

# Create a DataFrame
df = pd.DataFrame({
    'Name': np.random.choice(names, size=20),
    'Age': np.random.randint(22, 60, size=20),
    'Department': np.random.choice(departments, size=20),
    'Salary': np.random.randint(40000, 120000, size=20),
    'YearsExperience': np.random.randint(0, 20, size=20)
})
df.head()
```

# üß™ PRACTICE QUESTIONS BELOW


# 1Ô∏è‚É£ Use .loc to select all rows where the Department is 'Tech'.
#    - Describe what .loc does and why it's useful for label-based filtering.


```python
df.loc[df['Department']=='Tech']

# .loc[] is a Pandas property(accessor) that returns a dataframe based on a (row-column) specification, primarily used for label-based filtering 
# as it allows accessing and slicing parts of a dataframe based on the row/column labels (or conditions applied to either of them).
```


# 2Ô∏è‚É£ Use .iloc to select the first 5 rows and the last two columns.
#    - What is the difference between .loc and .iloc?


```python
df.iloc[:5,-2:]

#.loc allows you to access rows and columns by their labels (label-based data selection), 
# whilst .iloc does the same by using integer positions instead (index-based filtering).
```


# 3Ô∏è‚É£ Map a new column called 'DeptCode' where:
#     'Sales' -> 1, 'Marketing' -> 2, 'HR' -> 3, 'Tech' -> 4.
#    - Use .map and explain what happens if a value is not mapped.


```python
encoding = {'Sales':1, 'Marketing':2, 'HR':3, 'Tech':4}
df['DeptCode'] = df['Department'].map(encoding)
df.head()

# If a value in df['Department'] is not found in the mapping dictionary, .map() will assign 'NaN' to that row in the new column (DeptCode).
# Workaround: using fillna() to assign a valid value to those not mapped.
df['DeptCode'] = df['Department'].map(encoding).fillna(0)
df.head()
```


# 4Ô∏è‚É£ Use .apply to calculate a new column 'Seniority' where:
#     - If YearsExperience > 10 ‚Üí 'Senior'
#     - If between 5-10 ‚Üí 'Mid-Level'
#     - Else ‚Üí 'Junior'
#    - Use a lambda function with apply.


```python
df['Seniority'] = df['YearsExperience'].apply( lambda x: 'Senior' if x > 10 else
                                                         'Mid-Level' if (x > 5) & (x < 10) else 'Junior')
df
```


# 5Ô∏è‚É£ Overwrite all salaries for employees with < 3 years of experience to 35000.
#    - Use boolean indexing with .loc to do this.


```python
min(df['Salary'])
min(df['YearsExperience'])

# label-indexing
df.loc[df['YearsExperience']<3,'Salary'] = 35000
df
```

```python
# position-indexing
rowIndexies = df[df['YearsExperience']<3].index
colIndex = df.columns.get_loc('Salary')

df.iloc[rowIndexies,colIndex] = 35000
df
```


# 1Ô∏è‚É£1Ô∏è‚É£ Drop all rows where Age is below 25.
#     - Explain whether this modifies the DataFrame in place or returns a copy.


```python
# First method: using Filtering:
df[df['Age']>=25]

# Second method: using .drop()
df.drop(df[df['Age']<25].index,axis=0)

# This will only modifies a copy of the dataframe
# In order to modify the DataFrame in place we add 'inplace=True'
df.drop(df[df['Age']<25].index,axis=0,inplace=True)
df.reset_index(drop=True)
```


# 1Ô∏è‚É£4Ô∏è‚É£ Use .apply to normalize the Salary column (min-max scaling between 0 and 1).
#     - Bonus: Write your own normalization function and pass it to apply.


```python
def min_max_normalization(x):
    return (x - df['Salary'].min()) / (df['Salary'].max() - df['Salary'].min())

df['Salary']= df['Salary'].apply(min_max_normalization)
df
```

```python
def cramers_v(x, y):
    """
    Calculate Cramér's V statistic for categorical-categorical association.
    """
    confusion_matrix = pd.crosstab(x, y)
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape

    # Bias correction
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))    
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2corr / min((kcorr-1), (rcorr-1)))


cramers_results = {}
for col in categoricals:
    cramers_results[col] = cramers_v(df[col], df['target'])

cramers_df = pd.DataFrame.from_dict(cramers_results, orient='index', columns=['Cramers_V'])
print(cramers_df.sort_values(by='Cramers_V', ascending=False))
```

**Assessing Pair Wise Relationships -> Cramer's V**

```python
categoricals = df.select_dtypes(include="object").columns

# Pairwise Cramér’s V
cramers_matrix = pd.DataFrame(
    np.zeros((len(categoricals), len(categoricals))),
    index=categoricals,
    columns=categoricals
)

for col1 in categoricals:
    for col2 in categoricals:
        if col1 == col2:
            cramers_matrix.loc[col1, col2] = 1.0  # self-correlation
        else:
            cramers_matrix.loc[col1, col2] = cramers_v(df[col1], df[col2])

plt.figure(figsize=(10,8))
sns.heatmap(cramers_matrix, annot=True, fmt=".2f", cmap="Blues")
plt.title("Cramer's V Correlation between Categorical Features")
plt.show()
```

> Since some features are some strongly Correlated pairs (more than ~70% multicolinearity), it's better to keep only one of each, Cramer's V between feature and target would act as a refrence to help choose which one to of the pair to keep 

> (e.g. `has-ring` & `ring-type` are highly correlated with ~87% multicolinearity, but `has-ring` is less correlated with target with only 6% colinearity compared to `ring-type` with 20% colinearity, hence `has-ring` will be the one to be dropped)

#### 4) Linear Preprocessor Object:
Handling:

> 'Missing Values' -> Imputer Obj

> 'Outliers' -> Custom-Made Winsorizer Obj 

> 'Categorical-features' -> Encoder Obj 

> 'Numerical-Scale' -> Scaler Obj

> 'Skewness' -> Transformer Obj

**Wisorizer Object:**

```python
class Winsorizer(BaseEstimator, TransformerMixin):
    def __init__(self, lower=0.01, upper=0.99):
        self.lower = lower
        self.upper = upper
    
    def fit(self, X, y=None):
        # storing percentiles from training data
        self.lower_bounds = np.percentile(X, self.lower * 100, axis=0)
        self.upper_bounds = np.percentile(X, self.upper * 100, axis=0)
        return self
    
    def transform(self, X):
        # applying capping
        return np.clip(X, self.lower_bounds, self.upper_bounds)
```

```python
num_features = ["cap-diameter", "stem-width", "stem-height"]
cat_features = df.select_dtypes(include=['object']).columns.tolist()

# Numerical pipeline
linear_num_pipeline = Pipeline([
    ("lin_imputer", SimpleImputer(strategy="median")),
    ("lin_winsor", Winsorizer(lower=0.01, upper=0.99)),       
    ("lin_scaler", RobustScaler()),                              
    ("lin_power", PowerTransformer(method='yeo-johnson')),    
])

# Categorical pipeline
linear_cat_pipeline = Pipeline([
    ("lin_imputer", SimpleImputer(strategy="most_frequent")),
    ("lin_encoder", OneHotEncoder(drop='first', handle_unknown="ignore", sparse_output=False)) 
])

# Full preprocessor 
linear_preprocessor = ColumnTransformer([
    ("lin_num", linear_num_pipeline, num_features),
    ("lin_cat", linear_cat_pipeline, cat_features)
])
```

#### 5) Tree-Based Preprocessor Object:
Handling:

> 'Missing Values' -> Imputer Obj

> 'Categorical-features' -> Encoder Obj 

> 'Outliers' -> OPTIONAL

```python
# Numerical pipeline 
tree_num_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("tree_winsor", Winsorizer(lower=0.01, upper=0.99)) 
])

# Categorical pipeline
tree_cat_pipeline = Pipeline([
    ("tree_imputer", SimpleImputer(strategy="most_frequent")),
    ("tree_encoder", OrdinalEncoder(handle_unknown="use_encoded_value", unknown_value=-1))
])

tree_preprocessor = ColumnTransformer([
    ("tree_num", tree_num_pipeline, num_features),
    ("tree_cat", tree_cat_pipeline, cat_features)
])
```

### **Phase 3: Modelling & Tuning**

#### 6) Baseline model performance results:

```python
models = {
    'Logistic Regression': LogisticRegression(random_state=42, class_weight='balanced', max_iter=1000, solver='lbfgs'),
    'Decision Tree': DecisionTreeClassifier(random_state=42, class_weight='balanced'),
    'Random Forest': RandomForestClassifier(random_state=42, class_weight='balanced'),
    'SVM': LinearSVC(random_state=42, class_weight='balanced', max_iter=1000),
    'KNN': KNeighborsClassifier(),
    'Naive Bayes': GaussianNB()
}

models_needing_smote = ['KNN', 'Naive Bayes']
linear_models = ['Logistic Regression', 'SVM', 'KNN', 'Naive Bayes']

results = {} # To Store results of each classifier

#-----------------------------------------------------------------

for model_name, model in models.items():

    print(f"\n{'-'*50}")
    print(f"Training {model_name}")
    print(f"{'-'*50}")

    # Choosing appropriate preprocessor
    if model_name in linear_models:
        preprocessor = linear_preprocessor
    else:  
        preprocessor = tree_preprocessor

    # Pipeline with or without SMOTE
    if model_name in models_needing_smote:
        base_pipeline = ImbPipeline([
            ('preprocessor', preprocessor),
            ('smote', SMOTE(random_state=42)),
            ('classifier', model)
        ])
    else:
        base_pipeline = Pipeline([
            ('preprocessor', preprocessor),
            ('classifier', model)
        ])  

    # Training the model
    base_pipeline.fit(X_train, y_train)

    # Predictions
    y_pred = base_pipeline.predict(X_test)
    y_pred_proba = base_pipeline.predict_proba(X_test)[:, 1] if hasattr(base_pipeline, 'predict_proba') else None 

    # Evaluation
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))
    
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
    print()
    # Cross Validation
    try:
        cv_scores = cross_val_score(base_pipeline, X_train, y_train, cv=5, scoring='f1_weighted')
        print(f"Cross-validation F1 scores: {cv_scores}")
        print(f"Mean CV F1: {cv_scores.mean():.3f} (±{cv_scores.std():.3f})")
    except Exception as e:
        print(f"Cross-validation failed: {e}")

    # Storing Results
    results[model_name] = {
        'pipeline': base_pipeline,
        'y_pred': y_pred,
        'y_pred_proba': y_pred_proba,
        'accuracy': accuracy_score(y_test, y_pred),
        'report': classification_report(y_test, y_pred, output_dict=True),
    }
    
#---------------------------------------------------------------------------------
```

#### 7) Feature selection results and justification:

```python
# Feature selection strategies dictionary for different models

feature_selection_strategies = {
    'Logistic Regression': {
        'method': 'rfe',
        'estimator': LogisticRegression(random_state=42, class_weight='balanced', max_iter=1000, solver='lbfgs'),
        'n_features': 30
    },
    'Decision Tree': {
        'method': 'embedded',
        'use_importance': True,
        'estimator' : DecisionTreeClassifier(random_state=42, class_weight='balanced'),
        'threshold': 'mean'  # Keeping only features above mean importance
    },
    'Random Forest': {
        'method': 'embedded', 
        'use_importance': True,
        'estimator' : RandomForestClassifier(random_state=42, class_weight='balanced'),
        'threshold': 0.01  # Keeping features with importance > 0.01
    },
    'SVM': {
        'method': 'rfe',
        'estimator': LinearSVC(random_state=42, class_weight='balanced', max_iter=1000),
        'n_features': 30
    },
    'KNN': {
        'method': 'filter',  
        'selector': SelectKBest(mutual_info_classif, k=30)
    },
    'Naive Bayes': {
        'method': 'filter',
        'selector': SelectKBest(f_classif, k=30)
    }
}
```

```python
# Integrating feature selection in pipeline
def feature_selection_model_pipeline(model_name, model, preprocessor):
    
    strategy = feature_selection_strategies[model_name]
    
    if strategy['method'] == 'rfe':
        selector = RFE(
            estimator=strategy['estimator'],
            n_features_to_select=strategy['n_features']
        )
    elif strategy['method'] == 'embedded':
        selector = SelectFromModel(
        estimator=strategy['estimator'],
        threshold=strategy['threshold']
    )
    else: 
        selector = strategy['selector']
    
    # Pipeline
    if model_name in models_needing_smote:
        return ImbPipeline([
            ('preprocessor', preprocessor),
            ('feature_selector', selector),
            ('smote', SMOTE(random_state=42)),
            ('classifier', model)
        ])
    else:
        return Pipeline([
            ('preprocessor', preprocessor),
            ('feature_selector', selector),
            ('classifier', model)
        ])
```

```python
# Different tuning strategies for different models
tuning_strategies = {
    'Logistic Regression': {
        'method': 'grid',
        'params': {
            'classifier__C': [0.001, 0.01, 0.1, 1, 10, 100, 1000],
            'classifier__solver': ['lbfgs'],  # Simple and effective
            'classifier__penalty': ['l2'],    # Only compatible penalty
            'classifier__max_iter': [1000, 2000]
        },
        'n_iter': 10
    },
    'Decision Tree': {
        'method': 'grid',
        'params': {
            'classifier__max_depth': [None, 10, 20, 30, 50],
            'classifier__min_samples_split': [2, 5, 10],
            'classifier__min_samples_leaf': [1, 2, 4],
            'classifier__max_features': ['sqrt', 'log2', None]
        }
    },
    'Random Forest': {
        'method': 'random',
        'params': {
            'classifier__n_estimators': [200],
            'classifier__max_depth': [None],
            'classifier__min_samples_split': [5, 10],
            'classifier__min_samples_leaf': [1],
            'classifier__max_features': ['sqrt', 'log2'],
            'classifier__bootstrap': [False]
        },

    },
    'SVM': {
    'method': 'grid',
    'params': {
        'classifier__C': [10, 15, 20],  
        'classifier__loss': ['hinge'],          
        'classifier__tol': [1e-3, 1e-4, 1e-5],   
        'classifier__dual': [True],
        'classifier__max_iter': [10000]        
        },
    },
    'KNN': {
        'method': 'grid',
        'params': {
            'classifier__n_neighbors': [3, 5, 7],
            'classifier__weights': ['distance'],
            'classifier__p': [1] 
        }
    },
    'Naive Bayes': {
        'method': 'grid',
        'params': {
            'classifier__var_smoothing': [1e-9, 1e-8, 1e-7, 1e-6, 1e-5]
        }
    }
}
```

```python
def tune_model(model_name, pipeline, X_train, y_train):

    strategy = tuning_strategies[model_name]
    
    if strategy['method'] == 'grid':
        tuner = GridSearchCV(
            pipeline,
            strategy['params'],
            cv=3,  # Fewer folds for speed
            scoring='f1_weighted',
            n_jobs=-1, # to utilize all CPU cores
            verbose=1
        )
    elif strategy['method'] == 'random':
        tuner = RandomizedSearchCV(
            pipeline,
            strategy['params'],
            cv=3,
            n_iter= 4,
            scoring='f1_weighted',
            n_jobs=-1,
            verbose=1,
            random_state=42
        )
    
    print(f"Tuning {model_name}...")
    tuner.fit(X_train, y_train)
    
    return tuner.best_estimator_, tuner.best_params_, tuner.best_score_
```

```python
tuned_results = {}
best_params = {}

for model_name in models.keys():
    print(f"\n{'='*60}")
    print(f"Tuning {model_name}")
    print(f"{'='*60}")
    
    if model_name in linear_models:
        preprocessor = linear_preprocessor
    else:
        preprocessor = tree_preprocessor
    
    # Create pipeline (without feature selection since it hurt performance)
    if model_name in models_needing_smote:
        tuned_pipeline = ImbPipeline([
            ('preprocessor', preprocessor),
            ('smote', SMOTE(random_state=42)),
            ('classifier', models[model_name])
        ])
    else:
        tuned_pipeline = Pipeline([
            ('preprocessor', preprocessor),
            ('classifier', models[model_name])
        ])
    
    # Tuning the model
    try:
        best_estimator, best_param, best_score = tune_model(model_name, tuned_pipeline, X_train, y_train)
        
        # Evaluating on test set
        y_pred_tuned = best_estimator.predict(X_test)
        y_pred_proba_tuned = best_estimator.predict_proba(X_test)[:, 1] if hasattr(best_estimator, 'predict_proba') else None
        
        tuned_results[model_name] = {
            'estimator': best_estimator,
            'best_params': best_param,
            'best_cv_score': best_score,
            'test_accuracy': accuracy_score(y_test, y_pred_tuned),
            'test_f1': f1_score(y_test, y_pred_tuned, average='weighted'),
            'y_pred': y_pred_tuned,
            'y_pred_proba': y_pred_proba_tuned,
            'report': classification_report(y_test, y_pred_tuned, output_dict=True),
        }
        
        best_params[model_name] = best_param
        
        print(f" {model_name} tuning completed")
        print(f"   Best CV Score: {best_score:.4f}")
        print(f"   Test Accuracy: {tuned_results[model_name]['test_accuracy']:.4f}")
        
    except Exception as e:

        print(f"{model_name} tuning failed: {e}")
        
        # Using untuned model as fallback
        base_pipeline.fit(X_train, y_train)
        y_pred = base_pipeline.predict(X_test)
        tuned_results[model_name] = {
            'estimator': base_pipeline,
            'best_params': 'Tuning failed',
            'test_accuracy': accuracy_score(y_test, y_pred),
            'test_f1': f1_score(y_test, y_pred, average='weighted'),
            'report': classification_report(y_test, y_pred, output_dict=True),
        }
```

##### **2. ROC-AUC Curve Comparison:**

```python
def plot_all_roc_curves(results_dicts, titles, y_test):

    n_sets = len(results_dicts)
    fig, axes = plt.subplots(1, n_sets, figsize=(6*n_sets, 6), sharey=True)

    for ax, results_dict, title in zip(axes, results_dicts, titles):
        colors = plt.colormaps.get_cmap("Set3")(np.linspace(0, 1, len(results_dict)))

        for (model_name, results), color in zip(results_dict.items(), colors):
            if 'y_pred_proba' in results and results['y_pred_proba'] is not None:
                fpr, tpr, _ = roc_curve(y_test, results['y_pred_proba'])
                roc_auc = auc(fpr, tpr)

                ax.plot(fpr, tpr, color=color, lw=2,
                        label=f'{model_name} (AUC = {roc_auc:.3f})')

        # Reference line (Random Guess)
        ax.plot([0, 1], [0, 1], 'k--', lw=1.5)

        # Formatting
        ax.set_xlim([0.0, 1.0])
        ax.set_ylim([0.0, 1.05])
        ax.set_xlabel("False Positive Rate", fontsize=12)
        ax.set_ylabel("True Positive Rate", fontsize=12 if ax == axes[0] else 0)  
        ax.set_title(title, fontsize=14, fontweight="bold")
        ax.legend(loc="lower right", fontsize=9)
        ax.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()


# Usage
plot_all_roc_curves(
    results_dicts=[results, fs_results, tuned_results],
    titles=["Baseline Models", "With Feature Selection", "Tuned Models"],
    y_test=y_test
)
```

* FS generally hurts performance for almost all models
> Suggests that all or most features are informative; aggressive filtering removes useful signals.

2. Tuning effects:

* For top models (Random Forest, KNN), tuning doesn’t add much (they’re already near-perfect)
